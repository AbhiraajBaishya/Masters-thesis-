```python
MAX_PF = 100
MAX_TRACK = 60
MAX_SV = 10# Global Jet Features
jet_features = [
    'npv', 'ntrueInt', 'rho', 'fj_pt', 'fj_eta', 'fj_phi', 'fj_mass', 'fj_sdmass', 
    'fj_tau1', 'fj_tau2', 'fj_tau3', 'fj_tau21', 'fj_tau32', 'fj_doubleb', 
    'fj_ptDR', 'fj_relptdiff', 'fj_sdn2', 'fj_z_ratio', 'fj_nSV', 'fj_jetNTracks',
    'fj_trackSipdSig_0', 'fj_trackSipdSig_1', 'fj_trackSipdSig_2', 'fj_trackSipdSig_3',
    'fj_trackSipdSig_0_0', 'fj_trackSipdSig_0_1', 'fj_trackSipdSig_1_0', 'fj_trackSipdSig_1_1',
    'fj_trackSip2dSigAboveCharm_0', 'fj_trackSip2dSigAboveBottom_0', 'fj_trackSip2dSigAboveBottom_1',
    'fj_tau0_trackEtaRel_0', 'fj_tau0_trackEtaRel_1', 'fj_tau0_trackEtaRel_2',
    'fj_tau1_trackEtaRel_0', 'fj_tau1_trackEtaRel_1', 'fj_tau1_trackEtaRel_2',
    'fj_tau_vertexMass_0', 'fj_tau_vertexMass_1', 'fj_tau_vertexEnergyRatio_0', 
    'fj_tau_vertexEnergyRatio_1', 'fj_tau_flightDistance2dSig_0', 'fj_tau_flightDistance2dSig_1',
    'fj_tau_vertexDeltaR_0'
]

# PF Candidate Features
pf_features = [
    'pfcand_ptrel', 'pfcand_erel', 'pfcand_phirel', 'pfcand_etarel', 'pfcand_deltaR', 
    'pfcand_drminsv', 'pfcand_drsubjet1', 'pfcand_drsubjet2', 'pfcand_dz', 'pfcand_dzsig', 
    'pfcand_dxy', 'pfcand_dxysig', 'pfcand_fromPV', 'pfcand_VTX_ass', 'pfcand_lostInnerHits', 
    'pfcand_puppiw', 'pfcand_hcalFrac', 'pfcand_charge', 'pfcand_isChargedHad', 
    'pfcand_isNeutralHad', 'pfcand_isGamma', 'pfcand_isEl', 'pfcand_isMu', 'pfcand_mass'
]

# Track Features
track_features = [
    'trackBTag_DeltaR', 'trackBTag_Eta', 'trackBTag_EtaRel', 
    'trackBTag_JetDistVal', 'trackBTag_Momentum'
]

# SV Features (Commonly used in DeepDoubleB/ParticleNet)
sv_features = [
    'sv_pt', 'sv_deltaR', 'sv_mass', 'sv_ntracks', 'sv_chi2', 'sv_normchi2', 
    'sv_dxy', 'sv_dxysig', 'sv_d3d', 'sv_d3dsig', 'sv_costhetasvpv'
]
```

```python
spectators = ['fj_sdmass', 'fj_pt']
```

```python
def pad_sequence(array, max_len):
    """Pads or truncates an awkward array sequence to a fixed length."""
    return ak.fill_none(ak.pad_none(array, max_len, clip=True), 0)



import gc

def load_data(directory, max_files=None):
    files = sorted(glob.glob(os.path.expanduser(directory) + "*.root"))
    if max_files:
        files = files[:max_files]

    # Padding limits for different collections
    MAX_PF = 100 
    MAX_TRK = 60
    MAX_SV = 10
    
    # Storage lists
    jet_list, pf_list, trk_list, sv_list, label_list = [], [], [], [], []

    for f in files:
        if "merged_63.root" in f: continue
            
        print(f"Reading: {f}...", end=" ", flush=True)
        try:
            with uproot.open(f) as file_handle:
                tree = file_handle["deepntuplizer/tree"]
                
                # Load ALL feature groups
                branches = (jet_features + pf_features + track_features + 
                           sv_features + spectators +
                           ['sample_isQCD', 'fj_isQCD', 'fj_isH', 'fj_isBB'])
                data = tree.arrays(branches, library='ak')

                # Standard filtering
                mask = (data['fj_sdmass'] > 40) & (data['fj_sdmass'] < 200) & \
                       (data['fj_pt'] > 350) & (data['fj_pt'] < 2000)
                
                label_hbb = data['fj_isH'] * data['fj_isBB']
                label_qcd = data['fj_isQCD'] * data['sample_isQCD']
                valid = mask & ((label_qcd == 1) | (label_hbb == 1))
                
                filtered = data[valid]
                if len(filtered) == 0:
                    print("Empty.")
                    continue

                # 1. Labels and Global Jets (Standard NumPy)
                y = ak.to_numpy(ak.where(label_hbb[valid] == 1, 1, 0)).astype(np.float32)
                jets_np = np.stack([ak.to_numpy(filtered[f]) for f in jet_features], axis=-1).astype(np.float32)
                
                # 2. PF Candidates (Pad to 100)
                pf_p = ak.fill_none(ak.pad_none(filtered[pf_features], MAX_PF, axis=1, clip=True), 0)
                pfs_np = np.stack([ak.to_numpy(pf_p[f]) for f in pf_features], axis=-1).astype(np.float32)

                # 3. Tracks (Pad to 60)
                trk_p = ak.fill_none(ak.pad_none(filtered[track_features], MAX_TRK, axis=1, clip=True), 0)
                trks_np = np.stack([ak.to_numpy(trk_p[f]) for f in track_features], axis=-1).astype(np.float32)

                # 4. Secondary Vertices (Pad to 10)
                sv_p = ak.fill_none(ak.pad_none(filtered[sv_features], MAX_SV, axis=1, clip=True), 0)
                svs_np = np.stack([ak.to_numpy(sv_p[f]) for f in sv_features], axis=-1).astype(np.float32)

                # Append to lists
                jet_list.append(jets_np); pf_list.append(pfs_np)
                trk_list.append(trks_np); sv_list.append(svs_np)
                label_list.append(y)
                
                print(f"Done ({len(y)} jets)")
                del data, filtered, pf_p, trk_p, sv_p
                gc.collect() 
                    
        except Exception as e:
            print(f"\n[ERROR] {f}: {e}")
            continue

    return (np.concatenate(jet_list), np.concatenate(pf_list), 
            np.concatenate(trk_list), np.concatenate(sv_list), 
            np.concatenate(label_list))
```

```python
class ParticleDataset(Dataset):
    def __init__(self, jets, pfs, tracks, svs, labels):
        self.jets = torch.from_numpy(jets)
        self.pfs = torch.from_numpy(pfs)
        self.tracks = torch.from_numpy(tracks)
        self.svs = torch.from_numpy(svs)
        self.labels = torch.from_numpy(labels)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.pfs[idx], self.tracks[idx], self.svs[idx], self.jets[idx], self.labels[idx]
```

```python
class MultiInputTransformer(nn.Module):
    def __init__(self, pf_dim, tk_dim, sv_dim, jet_dim, embed_dim=128):
        super().__init__()
        # Sequence Encoders
        self.pf_enc = nn.Linear(pf_dim, embed_dim)
        self.tk_enc = nn.Linear(tk_dim, embed_dim)
        self.sv_enc = nn.Linear(sv_dim, embed_dim)
        
        # Shared Transformer architecture
        # We use batch_first=True because our data is (Batch, Sequence, Feature)
        encoder_layer = nn.TransformerEncoderLayer(d_model=embed_dim, nhead=8, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=3)
        
        # Jet Feature MLP
        self.jet_mlp = nn.Sequential(
            nn.Linear(jet_dim, embed_dim), 
            nn.ReLU(), 
            nn.Linear(embed_dim, embed_dim)
        )
        
        # Final Head: Now expects 4 sets of features concatenated (embed_dim * 4)
        self.classifier = nn.Sequential(
            nn.Linear(embed_dim * 4, 256),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, pfs, tracks, svs, jets):
        # 1. Process each sequence through the SAME transformer
        # Global Average Pooling (.mean(dim=1)) turns sequences into single vectors
        z_pf = self.transformer(self.pf_enc(pfs)).mean(dim=1)
        z_tk = self.transformer(self.tk_enc(tracks)).mean(dim=1)
        z_sv = self.transformer(self.sv_enc(svs)).mean(dim=1)
        
        # 2. Process Global Jet features with the MLP
        z_jet = self.jet_mlp(jets)
        
        # 3. Combine all 4 feature vectors
        combined = torch.cat([z_pf, z_tk, z_sv, z_jet], dim=-1)
        
        # 4. Classify
        return self.classifier(combined).squeeze()
```

```python
def train(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for pfs, tks, svs, jets, y in loader:
        # Move ALL inputs to the GPU
        pfs = pfs.to(device)
        tks = tks.to(device)
        svs = svs.to(device)
        jets = jets.to(device)
        y = y.to(device)
        
        optimizer.zero_grad()
        out = model(pfs, tks, svs, jets)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        # Add a print statement so you know it's working
        print(f".", end="", flush=True) 
        
    return total_loss / len(loader)

# %%
def evaluate(model, loader, device):
    model.eval()
    y_true, y_score, jet_pt = [], [], []
    
    print("Running evaluation for plotting...")
    with torch.no_grad():
        for pfs, tks, svs, jets, y in loader:
            pfs, tks, svs, jets, y = pfs.to(device), tks.to(device), svs.to(device), jets.to(device), y.to(device)
            out = model(pfs, tks, svs, jets)
            
            y_score.extend(out.cpu().numpy())
            y_true.extend(y.cpu().numpy())
            # Assuming fj_pt is index 0 in your jet_features list
            jet_pt.extend(jets[:, 0].cpu().numpy()) 
            
    y_true = np.array(y_true)
    y_score = np.array(y_score)
    jet_pt = np.array(jet_pt)

    # --- FIGURE 1: OVERALL ROC ---
    plt.figure(figsize=(8, 7))
    fpr_all, tpr_all, _ = roc_curve(y_true, y_score)
    roc_auc_all = auc(fpr_all, tpr_all)
    
    plt.plot(tpr_all, 1 - fpr_all, color='darkorange', lw=2, 
             label=f'Overall ROC (AUC = {roc_auc_all:.4f})')
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Tagging efficiency ($H \\rightarrow b\\bar{b}$)', fontsize=12)
    plt.ylabel('1 $-$ mistagging efficiency ($g \\rightarrow b\\bar{b}$)', fontsize=12)
    plt.title('Overall Tagger Performance', fontsize=14)
    plt.legend(loc="lower left")
    plt.grid(True, linestyle='--', alpha=0.6)
    
    plt.savefig('overall_roc_curve.png', dpi=300)
    print("Saved: overall_roc_curve.png")
    plt.show()

    # --- FIGURE 2: pT BINNED ROC ---
    plt.figure(figsize=(8, 7))
    
    # Plot 'All' as a reference
    plt.plot(tpr_all, 1 - fpr_all, label=f'All (AUC: {roc_auc_all:.3f})', 
             color='green', linestyle='-.', linewidth=2)

    # Define bins based on your figure
    pt_bins = [
        (300, 600, 'red', '--'),
        (600, 800, 'blue', ':'),
        (800, 2000, 'black', '-')
    ]

    for pt_min, pt_max, color, style in pt_bins:
        mask = (jet_pt >= pt_min) & (jet_pt < pt_max)
        if np.sum(mask) > 50: # Only plot if we have enough samples
            fpr, tpr, _ = roc_curve(y_true[mask], y_score[mask])
            auc_val = auc(fpr, tpr)
            label = f'{pt_min} < $p_T$ < {pt_max} GeV (AUC: {auc_val:.3f})'
            plt.plot(tpr, 1 - fpr, label=label, color=color, linestyle=style)

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Tagging efficiency ($H \\rightarrow b\\bar{b}$)', fontsize=12)
    plt.ylabel('1 $-$ mistagging efficiency ($g \\rightarrow b\\bar{b}$)', fontsize=12)
    plt.title('Tagger Performance by $p_T$ Bin', fontsize=14)
    plt.legend(loc="lower left", fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.6)
    
    plt.savefig('pt_binned_roc_curves.png', dpi=300)
    print("Saved: pt_binned_roc_curves.png")
    plt.show()
```

```python
# 1. Load data with all features
# I recommend setting max_files to 30 or 40 initially to avoid memory issues on Mac
train_jets, train_pfs, train_trks, train_svs, train_labels = load_data(train_directory, max_files=10)
test_jets, test_pfs, test_trks, test_svs, test_labels = load_data(test_directory, max_files=5)

# 2. Create PyTorch Datasets with all 4 feature types
# Make sure your ParticleDataset class matches the 5-input __init__ we defined
train_dataset = ParticleDataset(train_jets, train_pfs, train_trks, train_svs, train_labels)
test_dataset = ParticleDataset(test_jets, test_pfs, test_trks, test_svs, test_labels)

# 3. Create DataLoaders for batching
# Note: On Mac, you usually don't need num_workers > 0 for this amount of data
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)

print(f"Data Loaded: {len(train_dataset)} training jets, {len(test_dataset)} testing jets.")
```

```python
# Determine dimensions
pf_in_dim = len(pf_features)
jet_in_dim = len(jet_features)
tk_in_dim = len(track_features)
sv_in_dim = len(sv_features)

# Initialize and move to GPU
model = MultiInputTransformer(pf_dim=pf_in_dim, tk_dim=tk_in_dim, sv_dim=sv_in_dim, jet_dim=jet_in_dim)
model.to(device) # CRITICAL: Move model parameters to GPU

# Define optimizer and loss
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# 2. Run Training for 10 epochs
for epoch in range(1, 11):
    loss = train(model, train_loader, optimizer, criterion, device)
    print(f"Epoch {epoch}/10 | Loss: {loss:.4f}")

# 3. Final Performance Plot
evaluate(model, test_loader, device)
```
